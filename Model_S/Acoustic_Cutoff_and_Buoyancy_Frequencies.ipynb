```python
# read in binary data
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys, os
sys.path.append(os.environ['raco'])
from common import dotdict, g_univ, inds_from_vals
f = open("Model_S.pkl", "rb")
di = dotdict(pickle.load(f))
f.close()
```

```python
# all right let's plot some buoyancy frequency
```

```python
plt.close()
plt.plot(di.rr, di.nsq)
plt.xlabel('r [cm]')
plt.ylabel('N^2')
plt.yscale('log')
plt.ylim(1e-8,1e-5)
```

```python
plt.close()
plt.plot(di.rr, di.nsq)
plt.xlabel('r [cm]')
plt.ylabel('N^2')
#plt.yscale('log')
plt.ylim(0,1e-5)
```

```python
# all right let's plot some acoustic cutoff frequency
```

```python
csq = di.dlnprs/di.dlnrho*di.prs/di.rho
Hprs = -1./di.dlnprs
omcasq = csq/Hprs**2

plt.close()
plt.plot(di.rr, omcasq)
plt.xlabel('r [cm]')
plt.ylabel('w_ca^2')
plt.yscale('log')
#plt.ylim(1e-8,1e-5)
```

```python
plt.close()
plt.plot(di.rr, omcasq)
plt.xlabel('r [cm]')
plt.ylabel('w_ca^2')
#plt.yscale('log')
plt.ylim(0,5e-5)
```

```python
# now the ratio
```

```python
plt.close()
plt.plot(di.rr, di.nsq/omcasq)
plt.xlabel('r [cm]')
plt.ylabel('N^2/w_ca^2')
#plt.yscale('log')
plt.ylim(0,0.3)
```

```python
#  plot just the sound speed
plt.close()
plt.plot(di.rr, np.sqrt(csq))
plt.xlabel('r [cm]')
plt.ylabel('|cs|')
plt.yscale('log')
#plt.ylim(1e-8,1e-5)
```

```python
#  plot solar sound crossing time, in minutes
plt.close()
plt.plot(di.rr, 2*di.rsun/np.sqrt(csq)/60.)
plt.xlabel('r [cm]')
plt.ylabel('sound crossing time (minutes)')
#plt.yscale('log')
plt.ylim(0,300)
```

```python
# finally, try plotting instead normalized buoyancy frequency
hprs = - 1./di.dlnprs
plt.close()
plt.plot(di.rr, di.nsq/di.grav*hprs)
plt.xlabel('r [cm]')
plt.ylabel('N^2 / (g/H_P)')
#plt.yscale('log')
plt.ylim(0,0.3)
```
